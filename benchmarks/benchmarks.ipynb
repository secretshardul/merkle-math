```python
import matplotlib.pyplot as plt
import pandas as pd

columns = ["i", "gas"]
df_abdk = pd.read_csv("abdk.csv", usecols=columns)
df_merkle = pd.read_csv("merkle.csv", usecols=columns)

plt.plot(df_abdk.i, df_abdk.gas, '-.', label="ABDK")
plt.plot(df_merkle.i, df_merkle.gas, label="Merkle")

plt.xlabel("x")
plt.ylabel("gas for log10 (x)")
plt.title('ABDK vs merkle log gas benchmark')

plt.legend()
plt.show()
```
